```python
# Plot fractional incorrect misclassifications for the best performing ML model (RANDOM FOREST)
plt.figure(figsize=(14, 10))

# CNN
plt.subplot(2, 2, 1)
cm = confusion_matrix(y_test, prediction_1_rf)
incorr_fraction = 1 - np.diag(cm) / np.sum(cm, axis=1)
plt.bar(np.arange(2), incorr_fraction)
plt.xlabel('True Label')
plt.ylabel('Fraction of incorrect predictions')
plt.title('Random forest misclassification for the First Step using CNN')

# VGG16
plt.subplot(2, 2, 2)
cm = confusion_matrix(y_test, prediction_1_rf_vgg16)
incorr_fraction = 1 - np.diag(cm) / np.sum(cm, axis=1)
plt.bar(np.arange(2), incorr_fraction)
plt.xlabel('True Label')
plt.ylabel('Fraction of incorrect predictions')
plt.title('Random forest misclassification for the First step using VGG16')

# ResNet
plt.subplot(2, 2, 3)
cm = confusion_matrix(y_test, prediction_1_rf_resnet)
incorr_fraction = 1 - np.diag(cm) / np.sum(cm, axis=1)
plt.bar(np.arange(2), incorr_fraction)
plt.xlabel('True Label')
plt.ylabel('First step Fraction of incorrect predictions')
plt.title('Random forest misclassification for the First Step using ResNet')

# DenseNet
plt.subplot(2, 2, 4)
cm = confusion_matrix(y_test, prediction_1_rf_densenet)
incorr_fraction = 1 - np.diag(cm) / np.sum(cm, axis=1)
plt.bar(np.arange(2), incorr_fraction)
plt.xlabel('True Label')
plt.ylabel('Fraction of incorrect predictions')
plt.title('Random forest misclassification for the First Step using DenseNet')

plt.tight_layout()
plt.show()
```

### OVERALL BALANCED MULTICLASS ACCURACY

```python
# Calculate the accuracy for each class (Benign, Melanoma, Seborrheic) and BMA for each model (Using Random Forest)
models = ['CNN', 'VGG16', 'ResNet', 'DenseNet']

for model_name in models:
    print(f"\nMetrics for {model_name} Model:")
    
    # First Step (Benign vs Others)
    if model_name == 'CNN':
        y_pred_bin = prediction_1_rf
    else:
        y_pred_bin = globals()[f"prediction_1_rf_{model_name.lower()}"]
    
    cm_first = confusion_matrix(y_test, y_pred_bin)
    TN = cm_first[0, 0]
    FP = cm_first[0, 1]
    FN = cm_first[1, 0]
    TP = cm_first[1, 1]
    
    benign_accuracy = TN / (TN + FP)
    print('The accuracy of the benign class in the first step is:', benign_accuracy)
    
    # Second Step (Melanoma vs Seborrheic)
    if model_name == 'CNN':
        y_pred_step2_bin = prediction_2_rf
    else:
        y_pred_step2_bin = globals()[f"prediction_2_rf_{model_name.lower()}"]
    
    cm_second = confusion_matrix(y_test_step2, y_pred_step2_bin)
    TN_2 = cm_second[0, 0]
    FP_2 = cm_second[0, 1]
    FN_2 = cm_second[1, 0]
    TP_2 = cm_second[1, 1]

    melanoma_accuracy = TN_2 / (TN_2 + FP_2)
    print('The accuracy of the melanoma class in the second step is:', melanoma_accuracy)

    seborrheic_accuracy = TP_2 / (TP_2 + FN_2)
    print('The accuracy of the seborrheic class in the second step is:', seborrheic_accuracy)
    
    # Calculate BMA
    bma = 1/3 * (benign_accuracy + melanoma_accuracy + seborrheic_accuracy)
    print('Balanced Model Accuracy (BMA):', bma)
```
